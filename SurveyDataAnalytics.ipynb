### Age Distribution

```python
age_plot=df['Q2'].value_counts().reindex(['18-21','22-24',
                                            '25-29','30-34',
                                            '35-39','40-44',
                                             '45-49','50-54',
                                            '55-59','60-69',
                                            '70-79','80+'])
f,ax=plt.subplots()

age_plot.plot(ax=ax,
            color='r')
age_plot.plot(kind='bar',
             ax=ax,
             figsize=(14,7),title="Age Distribution")
```
